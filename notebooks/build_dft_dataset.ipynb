```python
def parse_line_dynamically(line):
    import re

    # Split the line into segments
    segments = re.findall(r'\w+="[^"]*"|\w+=\S+', line)
    
    # Initialize the dictionary to store the key-value pairs
    parsed_dict = {}

    # Iterate over each segment
    for segment in segments:
        if '=' in segment:
            key, value = segment.split('=', 1)
            value = value.replace('"', '').replace("'", '')
            # Try to convert the value into an appropriate type
            try:
                # If the value contains spaces, it's likely a list of numbers
                if ' ' in value:
                    value = list(map(float, value.split()))
                else:
                    value = float(value)
            except ValueError:
                # If conversion fails, keep the value as a string
                pass
            parsed_dict[key] = value

    return parsed_dict

def read_xyz(fileobj, cutoff: float, pbc: bool = True):
    import ase.atoms
    import ase.geometry
    import ase.neighborlist
    
    # This function reads first all atoms and then yields based on the index.
    # Perfomance could be improved, but this serves as a simple reference.
    # It'd require more code to estimate the total number of images
    # without reading through the whole file (note: the number of atoms
    # can differ for every image).

    lines = fileobj.readlines()
    while len(lines) > 0:
        ignore_H = False
        ignore_C = False
        ignore_N = False
        ignore_O = False
        
        symbols = []
        positions = []
        chemical_shifts = []
        natoms = int(lines.pop(0))
        parsed_result = parse_line_dynamically(lines.pop(0))  # Comment line;
        if 'Lattice' in parsed_result:
            lattice = parsed_result.pop('Lattice')

            cell = np.array(lattice).reshape(3, 3)
            parsed_result['cell'] = cell[np.newaxis, ...]
        
        name = parsed_result.get('NAME')
        if name in H_NAMES:
            ignore_H = True
        if name in C_NAMES:
            ignore_C = True
        if name in N_NAMES:
            ignore_N = True
        if name in O_NAMES:
            ignore_O = True
        
        for _ in range(natoms):
            line = lines.pop(0)
            symbol, x, y, z, cs = line.split()[:5]
            symbol = symbol.lower().capitalize()
            symbols.append(symbol)
            positions.append([float(x), float(y), float(z)])
            if (symbol == 'H' and ignore_H) \
            or (symbol == 'C' and ignore_C) \
            or (symbol == 'N' and ignore_N) \
            or (symbol == 'O' and ignore_O):
                chemical_shifts.append([np.nan])
            else:
                chemical_shifts.append([float(cs)])
        atoms = ase.atoms.Atoms(symbols=symbols, positions=positions)
        pos = atoms.positions
        ds = {
            'pos': pos[np.newaxis, ...],
            'node_type': atoms.get_atomic_numbers(),
            'cs': np.array(chemical_shifts)[np.newaxis, ...],
        }
        if pbc:
            cell = ase.geometry.complete_cell(cell)
            first_idex, second_idex, edge_cell_shift = ase.neighborlist.primitive_neighbor_list(
                "ijS",
                (True,) * 3,
                cell,
                pos,
                cutoff=cutoff,
                self_interaction=False,  # we want edges from atom to itself in different periodic images!
                use_scaled_positions=False,
            )
            # Remove self-node edges (a node with itself)
            bad_edge = first_idex == second_idex
            bad_edge &= np.all(edge_cell_shift == 0, axis=1)
            keep_edge = ~bad_edge
            if not np.any(keep_edge):
                raise ValueError(
                    f"Every single atom has no neighbors within the cutoff r_max={cutoff} (after eliminating self edges, no edges remain in this system)"
                )
            first_idex = first_idex[keep_edge]
            second_idex = second_idex[keep_edge]

            edge_cell_shift = edge_cell_shift[keep_edge].astype(float)
            edge_index = np.vstack([first_idex, second_idex])

            # Check pbc consistency
            x = pos[edge_index]
            dist_vec = x[1] - x[0]
            z = dist_vec + np.einsum(
                    "ni,ij->nj",
                    edge_cell_shift,
                    cell,  # remove batch dimension
                )
            assert np.linalg.norm(z, axis=-1).max() < cutoff

            ds.update({
                'edge_index': edge_index[np.newaxis, ...],
                'edge_cell_shift': edge_cell_shift[np.newaxis, ...],
            })
        ds.update(parsed_result)
        yield ds

with open(inputfile, 'r') as fileobj:
    for ds in read_xyz(fileobj, cutoff=5.0, pbc=True):
        npz_filename = os.path.join(outputroot, f"{ds['NAME']}.npz")
        np.savez(npz_filename, **ds)
        print(f"Molecule {ds['NAME']} saved!")
```
